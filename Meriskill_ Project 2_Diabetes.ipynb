# Model Building

```python
y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
# Define a list of models to run
models = [
    ('KNN', KNeighborsClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Decision Tree', DecisionTreeClassifier()),
    ('SVM', SVC()),
]
```

```python
from sklearn.model_selection import cross_validate
from tabulate import tabulate
import numpy as np

# Initialize a list to store the results
results_table = []

# Loop through each model
for name, regressor in models:
    # Perform cross-validation
    cv = cross_validate(regressor, X, y, cv=5, scoring=["accuracy", "f1", "recall", "precision"])

    # Append results to the table
    results_table.append([name,
                          np.mean(cv["test_accuracy"]),
                          np.mean(cv["test_precision"]),
                          np.mean(cv["test_recall"]),
                          np.mean(cv["test_f1"])])

# Convert the results to a DataFrame
results_df = pd.DataFrame(results_table, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

# Display the results using tabulate
table_str = tabulate(results_df, headers='keys', tablefmt='pretty', showindex=False)

# Print the table
print(table_str)

with open('output_table_cross_validation.txt', 'w') as file:
    file.write(table_str)
```

Here we can see that logistic regression is the best model for our given data so we can use this model for further prediction.
